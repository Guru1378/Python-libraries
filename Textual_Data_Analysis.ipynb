```python
 import re
```

```python
def html_tag_remover(data):
  data=re.sub(r'<.*?>','',data)
  return data
```

```python
data["review"]=data["review"].apply(html_tag_remover)
```

```python
def remove_urls(data):
  data=re.sub(r'https?://\S+|www\.\S+','',data)
  return data
```

```python
data["review"]=data["review"].apply(remove_urls)
```

```python
def spelling_correction(data):
  data=TextBlob(data).correct().string
  return data
```

```python
data["review"]=data["review"].apply(spelling_correction)
```

```python
import string
```

```python
string.punctuation
```

```python
def remove_punctuations(text):
  for i in string.punctuation:
    if i in text:
      text=text.replace(i,"")
  return text
```

```python
data["review"]=data["review"].apply(remove_punctuations)
```
