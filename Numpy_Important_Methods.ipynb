```python
def fun(x,y):
  return 3*x**2+4*y**2
zz=fun(xx,yy)
zz
```

```python
import plotly.express as px
import plotly.graph_objects as go

fig=px.scatter_3d()
fig.add_trace(go.Scatter3d(x=xx,y=yy,z=zz))
fig.show()
```
